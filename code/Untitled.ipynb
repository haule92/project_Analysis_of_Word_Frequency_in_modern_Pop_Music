```python
from enchant.checker import SpellChecker

#     return False if ((len(errors) > 1) or len(quote.split()) < 3) else True

def is_in_english(quote):
    d = SpellChecker("en_US")
    d.set_text(quote)
    errors = [err.word for err in d]
    if len(errors) > 0:
        return False
    return True
```

```python
def coun_uniques(str):
    uniques = []
    for word in str.split():
        if word not in non_unique and len(word)>2:
            uniques.append(word)
    return list(set(uniques))

unique_words_in_lyric = []
unique_count = []

for i in range(df.shape[0]):
    unique_words_in_lyric.append(coun_uniques(df["clean_lyrics"][i]))
    
    
    
    
    
```

```python
for i in range(df.shape[0]):
    unique_count.append(len(coun_uniques(df["clean_lyrics"][i])))
```
